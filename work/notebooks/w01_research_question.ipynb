## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total pages:", len(df))
print("Declining pages:", df["trend_direction"].value_counts().get("down", 0))
print("Declining rate:", round((df["trend_direction"]=="down").mean(), 3))

corr = df["search_volume"].corr(df["impressions_90d"])
print("Correlation between search volume and impressions:", round(corr, 3))
```

The dataset includes **30,000 pages**, with approximately **54%** labeled as declining. The correlation between search volume and impressions is close to zero, indicating minimal linear relationship between these variables. This suggests that search volume alone is not a reliable indicator for identifying pages that require updates. These characteristics support the relevance of exploring content refresh prioritization as a structured decision problem.
